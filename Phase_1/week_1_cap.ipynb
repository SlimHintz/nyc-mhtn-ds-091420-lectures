The above method, though exhaustive, is unecessary. All i want to do is clean all of the columns that are of type object. There is almost no point in making a highly specialised tool for ds. It is far better to create novle solutionson the fly

```python
# Define a helper function to return a stripped version of the cell input
def column_cleaner(column):
    return column.strip()
```

```python
# Define a function that finds all columns in a data frame that are of type object
def object_column_finder(df):
    return [column for column in df.columns if df[column].dtype == 'O']
```

```python
# Define a function that uses the two helper functions to return a cleaned copy of the orginal dataframe
def frame_object_cleaner(df):
    clean_df = df.copy()
    object_columns = object_column_finder(df) # generate a list of object columns
    for column in object_columns: # iterate through a list of columns
        clean_df[column] = clean_df[column].apply(column_cleaner) # strip white space from columns
    return clean_df
```
